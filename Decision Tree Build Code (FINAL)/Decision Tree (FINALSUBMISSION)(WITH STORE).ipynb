```python
import pandas as pd
import numpy as np
from sklearn import datasets
import math
```

# CODE TO BUILD DECISION TREE 

```python
#CODE
#USES RECURSION
#STORES WHOLE TREE IN THE DICTIONARY


d3={}
def build_tree6(df, y, unused_features,level,sublevel):
     
        possible_values=set(y[0])
        entropy_class=0           #"entropy_class" STORES ENTROPY OF A CLASS        #6th line
        check_value=set(y[0])     #"check_value" STORES ALL DISTINCT ELEMENTS in OUTPUT "y"
        
        #Loop CALCULATES ENTROPY of a CLASS
        #STORES it in variable "entropy_class"
        for i in possible_values:
            m=len((y[y[0]==i]))/len(y)
            entropy_class+=(-1*m*math.log(m,2))
            
        #---------------------------------------------------------------------------------------------------------
        #IF-ELSE CONDITION PART
        #CHECKS IF SHOULD RETURN OR SPLIT THE FEATURE
        
            
        #LEAF NODE CONDITION 
        #IF "Unused_Features" is EMPTY && OUTPUT is left with JUST ONE DISTINCT VALUE
        #PURE NODE CONDITION
        if (len(unused_features)==0 and len(check_value)==1):
            print()
            print("Level=",level)
            
            possible_values=set(y[0])                                                
            for i in possible_values:                                                #PRINTS COUNT OF DISTINCT OUTPUTS
                if(len(possible_values)!=0):
                    print("Count of ",i,"=",len(y[0]))   
                s="Count of "+str(i)
            
            print("Current Entropy=",entropy_class)                                  #PRINTS CLASS_ENTROPY FOR THAT LEVEL
            print("Reached Leaf Node")
            print("--------------------------------------------------------------")
            print()
            
            #----------------------------------------------------------------------------------------------------------
            #GIVING IN VALUES TO THE DICTIONARY
            
            if(level in d3):
                d3[level][sublevel]={}
            else:
                d3[level]={}
                d3[level][sublevel]={}
            d3[level][sublevel]["Entropy"]=entropy_class
            d3[level][sublevel]["Leaf Node"]="Yes"
            #-----------------------------------------------------------------------------------------------------------
            
            return
        
        #RETURN if PURE NODE CONDITION FULLFILLS ELSE
        #CALCULATE MAX_GAIN_RATION and SPLIT FEATURE with MAX_GAIN_RATION  
        else:
            
            Gain=0           #STORES DIFFERENCE of ENTROPY_CLASS AND ENTROPY_ATTRIBUTE i.e. GAIN
            Best_Feature=""
            max_gain=0       #STORES MAX_GAIN_RATIO out of all the FEATURES
            entropy=0        #STORES ENTROPY of a FEATURE
            
            #---------------------------------------------------------------------------------------------------------
            #1st LOOP 
            #LOOPS over all the "FEATURES" in "UNUSED_FEATURES"
            
            for features in unused_features:
                
                val=set(df[features])
                infogain=[]
                sum_all=0
                sum_counts=[]
                
                #---------------------------------------------------------------------------------------------------------
                #2nd LOOP
                #LOOPS over all DISTINCT VALUES in a FEATURE
                for v in val:
                    
                    #Get all distict values of column y
                    yval=set(y[0])
                    count=[]
                    sumy=0
                    
                    
                    #---------------------------------------------------------------------------------------------------------
                    #LOOP CALCULATES "SUM" of ALL DISTINCT VALUES IN OUTPUT of DISTINCT VALUE OF FAETURE
                    #SUPPOSE no of 0's in "a" of feature "sl_labeled"
                    #AND STORES NUMBER OF 0's,1's,2's IN COUNT[]
                    
                    for i in yval:
                        if len(yval)!=0:#54
                            t=y[df[features]==v]
                            count.append(len(t[t[0]==i]))
                            #ABOVE TWO LINES ARE FINDING NO. of 0,1,2 in (a,b,c,d)
                            sumy+=len(t[t[0]==i])
                    
                    sum_all+=sumy                    #FINDS P+N and stores IN sum_all
                    sum_counts.append(sum(count))    #FINDS Pi+Ni and stores IN sum_count[]
                    
                    #---------------------------------------------------------------------------------------------------------
                    #LOOP CALCULATES INFO GAIN FOR v IN (a,b,c,d)
                    #i.e. INFO GAIN FOR EVERY DISTINCT VALUE in FEATURE
                    
                    temp=0
                    for i in count:
                        if i!=0.0:
                            temp+=(-1*math.log((i/sumy),2)*(i/sumy))
                        else:
                            temp+=0
                    #STORES VALUE OF INFOGAIN OF v IN (a,b,c,d) in INFOGAIN[]
                    #USES THIS FOR CALCULATING ENTROPY of the FEATURE
                    infogain.append(temp)
                    
                    
                #---------------------------------------------------------------------------------------------------------    
                #LOOP CALCULATES ENTROPY FOR THE FEATURE
                p=0
                entropy=0
                for i in infogain:
                    #CONDITION AVOIDS A SITUATION where division is by 0
                    if(sum_all!=0):
                        entropy+=(i*sum_counts[p]/sum_all)
                    else:
                        entropy+=0
                    p+=1
                
                #---------------------------------------------------------------------------------------------------------
                #CALCULATING FEATURE GAIN AND BEST FEATURE AND BEST GAIN OUT OF UNUSED_FEATURES    
                Feature_Gain=entropy_class-entropy
                if Feature_Gain>max_gain:
                    max_gain=Feature_Gain
                    Best_Feature=features
                    curr_entropy=entropy
                    
            #---------------------------------------------------------------------------------------------------------
            #GIVING IN VALUES TO DICTIONARY
            
            if(level in d3):
                d3[level][sublevel]={}
            else:
                d3[level]={}
                d3[level][sublevel]={}
            
            d3[level][sublevel]["Level"]=level
            #----------------------------------------------------------------------------------------------------------
            #PRINTING PART
            
            
            print()
            print("Level =",level)                                                  #PRINTS LEVEL
            possible_values=set(y[0]) 
            
            for i in possible_values:                                                #PRINTS COUNT OF DISTINCT OUTPUTS
                if(len(possible_values)!=0):
                    print("Count of",i,"=",len(y[y[0]==i]))
                s="Count of "+str(i)
                d3[level][sublevel][s]=len(y[y[0]==i])
            
            print("Current Entropy=",entropy_class)                                  #PRINTS CLASS_ENTROPY FOR THAT LEVEL
            d3[level][sublevel]["Entropy"]=entropy_class
            
            if((len(check_value)>1))and(len(unused_features)>0):
                print("Splitting on Feature ",Best_Feature,"with Gain=",max_gain)
                d3[level][sublevel]['Best feature']=Best_Feature
                d3[level][sublevel]['Gain']=max_gain
                print("--------------------------------------------------------------")
                t_feature=df[Best_Feature]
                unused_feature_values=set(t_feature)
                dt=df.drop((Best_Feature),axis=1)
                
        
            #---------------------------------------------------------------------------------------------------------
            #RECURSION PART

                #LOOPS THROUGH ALL DISTINCT OUTPUT VALUES OF THE BEST_FEATURE
                #AND RECURSIVELY TRIES TO MAKE PATH FOR EACH DISTICT VALUE WITH 
                #REMAINING FEATURES IN "UNUSED_FEATURES"
                for k in unused_feature_values:
                    y1=y[df[Best_Feature]==k]
                    unused_features=set(dt.columns)
                    sublevel=k
                    build_tree6(dt,y1,unused_features,level+1,sublevel)#CALLED METHOD ITSELF #RECURSION
                    
            else:
                print("Reached Leaf Node")
                print("--------------------------------------------------------------")
            
            
            return 
```

```python
iris=datasets.load_iris()
df=pd.DataFrame(iris.data)
df.columns=['sl','sw','pl','pw']
```

```python
def label(k,*val):
    if k<val[0]:
        return 'a'
    elif k<val[1]:
        return 'b'
    elif k<val[2]:
        return 'c'
    else:
        return 'd'
```

```python
def toLabel(feature_name):
    minimum=df[feature_name].min()
    maximum=df[feature_name].max()
    second=df[feature_name].mean()
    first=(minimum+second)/2
    third=(maximum+second)/2
    return df[feature_name].apply(label,args=(first,second,third))  
```

```python
df['sl_labeled']=toLabel('sl')
df['sw_labeled']=toLabel('sw')
df['pl_labeled']=toLabel('pl')
df['pw_labeled']=toLabel('pw')
df.drop(['sl','sw','pl','pw'],axis=1,inplace=True)
```

# 2.) Testing on IRIS dataset

```python
y=pd.DataFrame(iris.target)
unused_features=set(df.columns)
level=0
sublevel=0
build_tree6(df,y,unused_features,level,sublevel)
```
